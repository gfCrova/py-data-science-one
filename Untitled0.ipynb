# Variables

```python
nombre = 'Gian'
edad = 28

print(f'Mi nombre es {nombre}, y tengo {edad} años!')
```

# Función con Input

```python
def saludar():
  nombre = input('Ingrese su nombre: ')
  print(f'{nombre} sea bienvenido/a!!!')
```

```python
saludar()
```

# Función Con Parámetros

```python
def saludar_con_parametros(a):
  print(f'{a} sea bienvenido/a!!!')
```

```python
saludar_con_parametros(nombre)
```

## **Calcular Velocidad Promedio**

```python
def velocidad(espacio, tiempo):
 v = espacio / tiempo
 print(f' Velocidad: {v} m/s')
```

```python
velocidad(100, 20)
```

# **Condicionales**

```python
def verificar_edad_para_conducir(edad):
  if edad >= 18:
    print('Usted se encuentra habilitado para conducir')
  else:
    print('Usted No tiene edad suficiente para conducir')
```

```python
verificar_edad_para_conducir(edad)
```
